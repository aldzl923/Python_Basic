```python
# 체인메소드 


print()
class Chain:
    def __init__(self, value=0):  #초기화
        self.value = value

    def add(self, x):
        self.value += x
        return self  # 자기 자신을 리턴하여 메소드 체인을 지원

    def multiply(self, y):
        self.value *= y
        return self

    def get_result(self):
        return self.value

# 객체 생성
obj = Chain()


# 메소드 체인을 통한 호출
result = obj.add(5).multiply(3).add(10).multiply(5).get_result()

print(result)  # 출력: 45 ( (0+5)*3+10 )
```

```python
class Item:  #클래스생성
    def __init__(self, Iprice=0, Itype=0): #초기화, 파라미터 정의, #Iprice, Itype default 0 으로 설정

        self.__Iprice = Iprice  #private 설정
        self.Itype = Itype

    def getIprice(self): #private으로 접근했기 때문에 그냥 접근이 불가함
        return self.__Iprice  #아이템 가격 변경못하게 하기 위해
    def getItype(self):
        return self.Itype
    

class Weapon(Item): ## 아이템을 상속한 웨폰이란 클래스 생성
    def __init__(self, Iprice=0, Itype=0, dropRate=0):   #Iprice, Itype default 0 으로 설정
        super().__init__(Iprice, Itype)  # Item 클래스의 생성자 호출, #자식클래스에서 부모클래스에 접근할때 super()을 사용
        self.__dropRate = dropRate # .__ : private로 설정 >> dropRate확률 조작하면 안됨

    def setDropRate(self, dropRate):
        self.__dropRate = dropRate
        
    def getDropRate(self):
        print(self.__dropRate)

myBow = Item(100, 'Bow')  # Item 클래스의 생성자와 Weapon 클래스의 생성자에 필요한 인자를 제공

print(f"Iprice = {myBow.getIprice()}, Itype = {myBow.getItype()}")
```

```python
import random

class Creature:
  def __init__(self, hp, mp, atk):
    self.hp = hp
    self.mp = mp
    self.atk = atk


  def attacked(self, enemy : 'Creature'):
    
    self.hp -= enemy.atk
    return self

    
class Player(Creature):
  def __init__(self, hp, mp, atk, critical_atk, avoidance):
      super().__init__(hp, mp, atk)
      self.critical_atk = critical_atk
      self.avoidance = avoidance
     

  
  def crticial_attacked(self, enemy : 'Creature'):
    self.hp -= enemy.atk * 2
    return self

  def avoided(self, enemy : 'Creature'):
    self.hp -= enemy.atk * 0
    return self
  
  
class Monster(Creature):
  def __init__(self, hp, mp, atk, stun, escape):
      super().__init(hp, mp, atk)
      self.stun = stun
      self.escape = escape
      
  def stuned(self, enemy : 'Creature'):
    self.hp = enemy.atk * 2
    return self

  def escaped(self, enemy : 'Creature'):
    self.hp = enemy.atk * 0
    return self
    

myPlayer = Player(20,10,5, True, False)
myMonster = Monster(5,3,1, True, False)

#myPlayer.attacked(myMonster)


#print(myPlayer.hp)

```

```python
import random

class pokemon:  
    def __init__(self, level, grade):
      
        self.level = level
        self.grade = grade
        
        
    def capture_probability(self, pokeball):
        base_probability = {'마스터볼': 0.5, '울트라볼': 0.4, '슈퍼볼': 0.25, '포켓볼': 0.1}.get(pokeball, 0)
        return base_probability + self.level * 0.05

    def catch_pokemon(self, pokeball):
        capture_prob = self.capture_probability(pokeball)
        return random.random() < capture_prob

player_level = 3
pokemon_grade = 'S'
pokeball_type = '포켓볼'

player_pokemon = pokemon(level=player_level, grade=pokemon_grade)

if player_pokemon.catch_pokemon(pokeball_type):
    print(f"당신은 {player_pokemon.level}레벨의 {player_pokemon.grade}등급 포켓몬을 잡았습니다!")
else:
    print("포켓몬을 놓쳤습니다.")
```
