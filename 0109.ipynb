```python
class Item:  #클래스생성
    def __init__(self, Iprice=None, Itype=None): #초기화, 파라미터 정의, #Iprice, Itype default 0 으로 설정

        self.__Iprice = Iprice
        self.Itype = Itype

    def getIprice(self): #private으로 접근했기 때문에 그냥 접근이 불가함
        return self.__Iprice  #아이템 가격 변경못하게 하기 위해

class Weapon(Item): ## 아이템을 상속한 웨폰이란 클래스 생성
    def __init__(self, Iprice=None, Itype=None, dropRate=0):   #Iprice, Itype default 0 으로 설정
        super().__init__(Iprice, Itype)  # Item 클래스의 생성자 호출, #자식클래스에서 부모클래스에 접근할때 super()을 사용
        self.__dropRate = dropRate # .__ : private로 설정 >> dropRate확률 조작하면 안됨

    def setDropRate(self, dropRate):
        self.__dropRate = dropRate
    def getDropRate(self):
        print(self.__dropRate)

myBow = Weapon(100, 'Bow', 0.01)  # Item 클래스의 생성자와 Weapon 클래스의 생성자에 필요한 인자를 제공

myBow.__dropRate = 20  ## 다른 변수(Weapon의 dropRate가 아님)

print(myBow.__dropRate)
myBow.getDropRate()
```

```python
import random

class pokemon:  
    def __init__(self, level, grade):
      
        self.level = level
        self.grade = grade
        
        
    def capture_probability(self, pokeball):
        base_probability = {'마스터볼': 0.5, '울트라볼': 0.4, '슈퍼볼': 0.25, '포켓볼': 0.1}.get(pokeball, 0)
        return base_probability + self.level * 0.05

    def catch_pokemon(self, pokeball):
        capture_prob = self.capture_probability(pokeball)
        return random.random() < capture_prob

player_level = 3
pokemon_grade = 'S'
pokeball_type = '포켓볼'

player_pokemon = pokemon(level=player_level, grade=pokemon_grade)

if player_pokemon.catch_pokemon(pokeball_type):
    print(f"당신은 {player_pokemon.level}레벨의 {player_pokemon.grade}등급 포켓몬을 잡았습니다!")
else:
    print("포켓몬을 놓쳤습니다.")
```

```python
import random

class game:
    def __init__(self, HP, stun_probability):  
        self.HP = HP
        self.stun_probability = stun_probability
        
        if self.HP <= 0:
            print("플레이어가 사망했습니다.")
            
class Player(game):
    def __init__(self, HP, Basic_atk, stun_atk):
        super().__init__(HP, stun_probability=0.5)
        self.Basic_atk = Basic_atk
        self.stun_atk = stun_atk
            
class Monster(game): 
    def __init__(self, HP, name, damage):
        super().__init__(HP, stun_probability=0.5)  ## stunprob 초기화 해야함
        self.name = name
        self.damage = damage
        
Golem = Monster(100, '골렘', 10)
print(Golem.HP)
print(Golem.damage)
```

```python
import random

class Creature:
    def __init__(self, hp, mp, atk):
        self.hp = hp
        self.mp = mp
        self.atk = atk

    def attacked(self, enemy: 'Creature'):
        self.hp -= enemy.atk
        return self


class Player(Creature):
    def __init__(self, hp, mp, atk, critical_atk, avoidance):
        super().__init__(hp, mp, atk)
        self.critical_atk = critical_atk
        self.avoidance = avoidance

    def critical_attacked(self, enemy: 'Creature'):
        self.hp -= enemy.atk * 2
        return self

    def avoided(self, enemy: 'Creature'):
        self.hp -= enemy.atk * 0
        return self


class Monster(Creature):
    def __init__(self, hp, mp, atk, stun, escape):
        super().__init__(hp, mp, atk)
        self.stun = stun
        self.escape = escape

    def stuned(self, enemy: 'Creature'):
        self.hp -= enemy.atk * 2
        return self

    def escaped(self, enemy: 'Creature'):
        self.hp -= enemy.atk * 0
        return self

myPlayer = Player(20, 10, 5, True, False)
myMonster = Monster(5, 3, 1, True, False)

myPlayer.attacked(myMonster)

print(myPlayer.hp)
```

```python
## 메이플 스토리 게임

class Unit:
    hp = None
    normal_attack_dmg = None

    def attack(self, enemy_hp):
        return enemy_hp - self.normal_attack_dmg


class Warrior(Unit):
    hp = 100
    normal_attack_dmg = 10
    power_strike_dmg = 20

    def power_strike(self, enemy_hp):
        print("파워 스트라이크!!!")
        return enemy_hp - self.power_strike_dmg

    def drink_potion(self):
        print("체력회복!!!")
        self.hp = self.hp + 10


class Slime(Unit):
    hp = 150
    normal_attack_dmg = 15
    body_attack_dmg = 25

    def body_attack(self, enemy_hp):
        print("몸통박치기!!!")
        return enemy_hp - self.body_attack_dmg


warrior = Warrior()
slime = Slime()

while True:
    print("슬라임이 공격합니다.")
    print("1. 일반공격")
    print("2. 몸통박치기")

    slime_input = input('사용할 스킬의 번호를 입력하세요: ')

    print("전사 체력:", warrior.hp)

    if slime_input == "1":
        warrior.hp = slime.attack(warrior.hp)
    elif slime_input == "2":
        warrior.hp = slime.body_attack(warrior.hp)

    print("전사 체력:", warrior.hp)

    if warrior.hp <= 0:
        break  # If warrior's HP is 0 or less, exit the loop.

    print("")

    print("전사가 공격합니다.")
    print("1. 일반공격")
    print("2. 파워스트라이크")
    print("3. 포션사용")

    warrior_input = input("사용할 스킬의 번호를 입력하세요: ")

    print("슬라임 체력:", slime.hp)

    if warrior_input == "1":
        slime.hp = warrior.attack(slime.hp)
    elif warrior_input == "2":
        slime.hp = warrior.power_strike(slime.hp)
    elif warrior_input == "3":
        warrior.drink_potion()
        print("전사 체력:", warrior.hp)

    print("슬라임 체력:", slime.hp)

    if slime.hp <= 0:
        break  # If slime's HP is 0 or less, exit the loop.

print("게임을 종료합니다.")
```
